```python
%matplotlib inline
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
import lasso
```

```python
#generate samples
np.random.seed(0)
X = np.arange(0, 10, 0.05)
y = np.sin(X) + 0.2 * X  + np.array([0.2*np.random.rand() for _ in range(len(X))])
```

```python
#サンプルをシャッフル
p = np.random.permutation(len(X))
X = X[p]
y = y[p]

train_ratio = 0.8
X_train, y_train, X_test, y_test = X[:len(X)*train_ratio],  y[:len(X)*train_ratio], X[len(X)*train_ratio:],  y[len(X)*train_ratio:]
```

```python
clf = lasso.GaussianKernelLinearRegressor(X_train, y_train, h = 0.2, l1 = 0.1, iter_num = 100)
```

```python
clf.test(X_test, y_test)
```
